## PLANET VISUALIZER

### ¿Cómo puedo usar el simulador? :

El simulador comienza a funcionar al correr este notebook, que utiliza el código del script 'simulation.py' dentro de la carpeta planet_viz_script creado por nosotros. La visualización de un sistema está definida por el parámetro que le entreguemos a la función simulation() que está en la última celda de este notebook (deslizar hacia abajo para revisar).

#### Hay 2 opciones principales para elegir :

- Si simulation() se ejecuta sin parametros de entrada, por defecto la función hará un random.choice y mostrará un sistema cualquiera de los disponibles en el data frame de forma aleatoria.

- Si se desea visualizar un sistema o planeta en particular, la función recibe el nombre de una host star perteneciente al dataframe y busca los planetas asociados a dicha estrella andfitriona junto a sus características físicas y ejecuta los cálculos necesarios para mostrar un accurate display de la órbita de los exoplanetas que brinden la información suficiente como para ser representados.
